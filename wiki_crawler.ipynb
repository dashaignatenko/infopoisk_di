```python
import re
import requests
```

```python
urls = ['https://mo.wikipedia.org/w/index.php?title=Special:Toate_paginile&from=1917', 'https://mo.wikipedia.org/w/index.php?title=Special:Toate_paginile&from=%D0%9B%D0%B8%D0%BC%D0%B1%D1%8D+%D0%BE%D1%84%D0%B8%D1%87%D0%B8%D0%B0%D0%BB%D1%8D']
```

```python
def articles(urls):
    artcls = []
    for i in urls:
        req = requests.get(i).text
        req = re.sub('<div id=\"catlinks".*</html>', '', req, flags=re.DOTALL)
        artcls += re.findall('<a href=\"(/wiki/[^"]*)\"[^>]*title=\"([^"]*)\">[^<]*</a></li>', req)
    return [('https://mo.wikipedia.org' + a[0],a[1]) for a in artcls]
```

```python
a = articles(urls)
```
