```python
def indices():
    d = {}
    for root, dirs, files in os.walk('.'):
        for file in files:
            if file.endswith('.txt'):
                f = open(file, 'r', encoding='utf-8')
                lemmas = m.lemmatize(f.read())
               # for i in set(lemmas):
                for i in lemmas:
                    if i not in stopw:
                        if i not in d:
                            d[i] = [file]
                        else:
                            d[i].append(file)
                f.close()
    return(d)
```
